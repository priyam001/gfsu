```python
#import 
import numpy as np        #it's a math librarey
import pandas as pd
from sklearn import linear_model         #Scikit Learn : https://scikit-learn.org/stable/user_guide.html
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score   # goodness of the model
```

```python
# Load the housing dataset
dataset = pd.read_csv('housing.csv')
dataset.head(10)
```

```python
#Get X and y from datset
X = dataset.iloc[:,:-1]    #iloc[]s used when the index label of a data frame is something other than numeric series of 0, 1, 2, 3â€¦.n
y = dataset.iloc[:,-1:]
X.fillna(0, inplace=True)
y.fillna(0, inplace=True)
print(X)
print(y)
```

```python
#split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
print(X_train, X_test, y_train, y_test)
```

```python
x=dataset.iloc[:,:-1:]
x
```

```python
#Create Linear Regression object
regr = linear_model.LinearRegression()
```

```python
#train the model
regr.fit(X_train, y_train)
```

```python
#test the model
y_pred = regr.predict(X_test)
```

```python
#check the goodness of fit using r2 score
print('Accuracy Score:', r2_score(y_test, y_pred)) 
```
